##### Exercício 7.5

```python
prompt = "\nQual a idade?"
prompt += "\n(Enter 'quit' when you are finished.) "
while True:
    idade = input(prompt)
    if idade == 'quit':
        break
    elif int(idade) < 3:
        print("O ingresso é gratuito.")
    elif int(idade) <= 12:
        print("O ingresso custará 10 dólares.")
    elif int(idade) > 12:
        print("O ingresso custará 15 dólares")
```

##### Exercício 7.8

```python
sandwich_orders = ['pastrami', 'mussarela', 'mortadela', 'frango', 'pastrami',
                   'presunto e queijo', 'picanha', 'linguiça', 'pastrami', 'vegano', ]
finished_sandwishes = []
print("Estamos sem pastrami!\n")
while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')
while sandwich_orders:
    current_sandwich = sandwich_orders.pop()
    print("Preparei seu sanduiche de " + current_sandwich + "!")
    finished_sandwishes.append(current_sandwich)
print("\nSanduiches entregues:")
for finished_sandwishe in finished_sandwishes:
    print(finished_sandwishe.title())
```

##### Exercício 7.10

```python
respostas = {}
prompt = "Se pudesse visitar um lugar do mundo, para onde você iria? "
while True:
    name = input("Qual o seu nome? ")
    lugar = input(prompt)
    respostas[name] = lugar
    repeat = input("Would you like to let another person respond?(yes/no) ")
    if repeat == 'no':
        break
print("\n---Poll Results---")
for name, lugar in respostas.items():
    print(name + " tem o sonho de visitar " + lugar + ".")
```
